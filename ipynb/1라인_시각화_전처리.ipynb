### 위 데이터로 folium으로 지도 시각화 진행하며 몇가지 문제점 및 개선점 확인

1. 위도, 경도로 표시한 곳과 실제 주소가 다름을 확인

2. '주소'의 데이터엔 구주소와 신주소가 혼재되어 있음

- 역지오코딩(위도, 경도로 주소 추출)으로 얻은 주소가 기존 주소와 상이하면 행 삭제
    * 카카오맵 API로 '위도', '경도'로 구주소가 들어있는 '상세주소' 칼럼 생성
    * '상세주소'의 앞 2글자가 '주소'에 있으면 정상 데이터로 판단
    * '주소'로 검색시 '상세주소'이용 -> 기존 OO구 까지만되던 검색이 OO동까지 가능



```python
API_KEY = '4fc49b4d8accc675a34999793db69f13'

# 좌표로 구주소(address_name)를 얻는 함수
def get_address_from_coords(lat, lng, api_key=API_KEY):
    url = "https://dapi.kakao.com/v2/local/geo/coord2address.json"
    headers = {"Authorization": f"KakaoAK {api_key}"}
    params = {
        "x": lng,  # 경도
        "y": lat,  # 위도
        "input_coord": "WGS84"
    }

    try:
        response = requests.get(url, headers=headers, params=params)
        result = response.json()
        documents = result.get('documents')

        # 응답에 주소 정보가 있으면 address_name 반환
        if documents:
            address = documents[0].get('address')
            if address:
                return address.get('address_name')

    # 예외 발생 시 None 반환 (좌표 오류, 응답 오류 등)
    except:
        return None

    return None  # documents가 비어있을 경우
```

```python
# 중복되는 위도, 경도 좌표만 추출하여 따로 저장
unique_coords = media_loc[['위도', '경도']].drop_duplicates().reset_index(drop=True)

# tqdm 진행률 적용
tqdm.pandas()

# 역지오코딩(위도, 경도로 주소 추출) 수행
unique_coords['상세주소'] = unique_coords.progress_apply(
    lambda row: get_address_from_coords(row['위도'], row['경도']), axis=1
)

# 원본 데이터프레임에 역지오코딩 결과 병합
media_loc = media_loc.merge(unique_coords, on=['위도', '경도'], how='left')
```

```python
# 혹시 모르니깐 카피
filtered_media_loc = media_loc.copy()
```

```python
filtered_media_loc[filtered_media_loc['상세주소'].str.contains('전남', na=False)]
```

```python
# '상세주소' 컬럼에서 결측값 제거 후, 각 주소의 앞 두 글자만 추출
first_two_chars = filtered_media_loc['상세주소'].dropna().apply(lambda x: x[:2])

# 중복을 제거한 고유한 앞 두 글자 리스트 생성
first_two_chars_list = first_two_chars.unique().tolist()

# 결과 출력
first_two_chars_list
```

```python
# 앞 두 글자 → 전체 시/도 이름 매핑
region_map = {
    '서울': '서울특별시',
    '부산': '부산광역시',
    '대구': '대구광역시',
    '인천': '인천광역시',
    '광주': '광주광역시',
    '대전': '대전광역시',
    '울산': '울산광역시',
    '세종': '세종특별자치시',
    '경기': '경기도',
    '강원': '강원특별자치도',
    '충북': '충청북도',
    '충남': '충청남도',
    '전북': '전라북도',
    '전남': '전라남도',
    '경북': '경상북도',
    '경남': '경상남도',
    '제주': '제주특별자치도',
}

# 상세주소 앞 2글자를 시/도로 치환
filtered_media_loc['상세주소'] = filtered_media_loc['상세주소'].apply(
    lambda x: region_map.get(x[:2], x[:2]) + x[2:] if isinstance(x, str) else x
)
```

```python
# 상세주소의 앞 2글자가 주소에 없으면 데이터 지우는 함수
def check_first_two_chars_in_address(row):
    상세주소 = str(row['상세주소'])
    주소 = str(row['주소'])
    first_word = 상세주소.split()[0] if 상세주소.split() else ''
    first_two_chars = first_word[:2]
    return first_two_chars in 주소

filtered_media_loc = filtered_media_loc[filtered_media_loc.apply(check_first_two_chars_in_address, axis=1)]
```

```python
# 전처리 된 데이터 저장
filtered_media_loc.to_csv("filtered_media_loc.csv", index = False)
```
